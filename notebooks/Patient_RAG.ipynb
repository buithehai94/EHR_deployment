# III. Embedding

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')

from tqdm import tqdm

combined_embeddings_dict = {}
print("Combining explanations and symptoms into embeddings with progress tracking...")
with tqdm(total=len(filtered_df), desc="Processing rows") as pbar:
    for i, (index, row) in enumerate(filtered_df.iterrows()):  # Correctly unpacking the iterrows()
        explanation_text = row['explanation']  # Accessing the row data directly
        symptoms_text = row['user']  # Accessing the row data directly
        
        explanation_embedding = model.encode(explanation_text, convert_to_tensor=True)
        symptom_embedding = model.encode([symptoms_text], convert_to_tensor=True)
        
        combined_embeddings_dict[i] = (explanation_embedding, symptom_embedding)
        pbar.update(1)  # Update progress for each row

print("Combined embeddings created successfully.")
```

```python
from tqdm import tqdm
import faiss
import numpy as np
import pickle

with open("combined_embeddings_dict.pkl", "rb") as f:
    combined_embeddings_dict = pickle.load(f)

print("Dictionary loaded successfully.")
```

```python
from sentence_transformers import SentenceTransformer
# Initialize model and variables
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
import numpy as np
# Step 1: Combine embeddings into numpy arrays
explanation_embeddings = []
symptom_embeddings = []

for idx in combined_embeddings_dict:
    explanation_embeddings.append(combined_embeddings_dict[idx][0].cpu().numpy())  # Convert to numpy
    symptom_embeddings.append(combined_embeddings_dict[idx][1].cpu().numpy())      # Convert to numpy

explanation_embeddings = np.vstack(explanation_embeddings)
symptom_embeddings = np.vstack(symptom_embeddings)

# Step 2: Create FAISS indexes
dimension = explanation_embeddings.shape[1]  # Embedding dimension
explanation_index = faiss.IndexFlatL2(dimension)  # L2 distance index
symptom_index = faiss.IndexFlatL2(dimension)

# Step 3: Add embeddings to the indexes
explanation_index.add(explanation_embeddings)
symptom_index.add(symptom_embeddings)

# Step 4: Save the FAISS indexes
faiss.write_index(explanation_index, "explanation_embeddings.index")
faiss.write_index(symptom_index, "symptom_embeddings.index")

print("FAISS indexes saved successfully.")
```

# Summarize information

```python
import requests
import json
from langchain_core.prompts import PromptTemplate

# Define the prompt template
TEMPLATE = "Here is the historical medication of patient: {historical_info}. \
            Summarize the patient information and answer whether the patient have any medical risk?"
prompt_template = PromptTemplate.from_template(TEMPLATE)
```

```python
# Endpoint URL
url = f"https://generativelanguage.googleapis.com/v1beta/models/gemini-1.5-flash-latest:generateContent?key={api_key}"

# Function to interact with the API using the prompt template
def chat(historical_info):
    # Format the template with the input `historical_info`
    formatted_prompt = prompt_template.format(historical_info=historical_info)

    # Payload for the API request
    payload = {
        "contents": [
            {
                "parts": [
                    {"text": formatted_prompt}
                ]
            }
        ]
    }

    # Headers for the request
    headers = {
        "Content-Type": "application/json"
    }

    # Send the request
    response = requests.post(url, headers=headers, data=json.dumps(payload))

    # Process and display the response
    if response.status_code == 200:
        response_json = response.json()
        candidates = response_json.get("candidates", [])
        if candidates:
            content_parts = candidates[0].get("content", {}).get("parts", [])
            if content_parts:
                for part in content_parts:
                    text = part.get("text", "")
                    formatted_text = text.replace("**", "")
                    print(formatted_text)
                    print("\n" + "-" * 80 + "\n")  # Add separators for readability
            else:
                print("No content parts found.")
        else:
            print("No candidates found in the response.")
    else:
        print(f"Error: {response.status_code}")
        print("Message:", response.text)
```
